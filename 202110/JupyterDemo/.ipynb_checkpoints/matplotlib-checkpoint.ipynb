# matplotlib

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,10,100)
x
```

```python
y = np.sin(x)
y
```

```python
plt.plot(x,y)
plt.show()
```

```python
#折线图
cosy = np.cos(x)
siny = y.copy()
plt.plot(x,siny,color="green",linestyle=":",label="sin(x)")
plt.plot(x,cosy,color="red",label="cos(x)")
#x轴的范围
plt.xlim(-5,10)
#对两轴的范围一起调整
plt.axis([-1,11,-2,2])
plt.xlabel("x label")
plt.ylabel("y label")
plt.legend()
plt.title("hello world")
plt.show()
```

```python
#散点图(二维特征)
x = np.random.normal(0,1,1000)
y = np.random.normal(0,1,1000)
plt.scatter(x,y,alpha=0.5)
plt.show()
```

```python
from sklearn import datasets
```

```python
iris = datasets.load_iris()
```

```python
iris.keys()
```

```python
X = iris.data
y = iris.target
plt.scatter(X[y==0,0],X[y==0,1],color="red",marker="o",label = iris.target_names[0])
plt.scatter(X[y==1,0],X[y==1,1],color="blue",marker="+",label = iris.target_names[1])
plt.scatter(X[y==2,0],X[y==2,1],color="green",marker="*",label = iris.target_names[2])
plt.legend()
plt.show()
```

```python
X = iris.data
y = iris.target
plt.scatter(X[y==0,2],X[y==0,3],color="red",marker="o",label = iris.target_names[0])
plt.scatter(X[y==1,2],X[y==1,3],color="blue",marker="+",label = iris.target_names[1])
plt.scatter(X[y==2,2],X[y==2,3],color="green",marker="*",label = iris.target_names[2])
plt.legend()
plt.show()
```
